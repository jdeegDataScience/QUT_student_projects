# 2) Which variables were included in your analysis and what were their roles and measurement level set? Justify your choice. 
Energy (Role: Feature, Measurement Level: Ratio):
Role: Energy is one of the key features used for clustering and analyzing audio tracks. It helps determine the intensity and activity of the tracks.
Measurement Level: Ratio level. Energy values range from 0.0 to 1.0 and represent a continuous scale with a true zero point. This allows for meaningful arithmetic operations (e.g., averaging, comparison).

Loudness (Role: Feature, Measurement Level: Ratio):
Role: Loudness is another important feature for clustering audio tracks, as it represents the overall volume or amplitude of the tracks.
Measurement Level: Ratio level. Loudness values are in decibels (dB), which are on a logarithmic scale with a true zero point. This makes loudness a ratio-level variable, allowing for meaningful mathematical operations.

Speechiness (Role: Feature, Measurement Level: Ratio):
Role: Speechiness indicates the presence of spoken words in a track. It can be useful for identifying tracks with spoken content.
Measurement Level: Ratio level. Speechiness values range from 0.0 to 1.0, representing the proportion of speech-like content in a track. It is a continuous variable.

Instrumentalness (Role: Feature, Measurement Level: Ratio):
Role: Instrumentalness indicates whether a track contains vocals or is instrumental. It can help in distinguishing instrumental tracks from vocal ones.
Measurement Level: Ratio level. Instrumentalness values range from 0.0 to 1.0, with higher values indicating a higher likelihood of instrumental content. It is a continuous variable.

Time Signature (Role: Feature, Measurement Level: Ordinal):
Role: Time signature represents the notational convention used in Western music to specify how many beats are contained in a measure. It can provide information about the rhythm and structure of the tracks.
Measurement Level: Ordinal level. Time signature values range from 0 to 5, and they represent categories or labels that indicate the time signature type. While they have an order, the intervals between them are not uniform or meaningful in a ratio sense.

ID (Role: Identifier, Measurement Level: Nominal):
Role: The 'ID' variable serves as a unique identifier for each audio track. It is not used as a feature for analysis but is essential for tracking and referencing specific tracks.
Measurement Level: Nominal level. 'ID' values are labels or identifiers without inherent order or magnitude. They are used for identification purposes.

Name (Role: Identifier, Measurement Level: Nominal):
Role: The 'Name' variable contains the names or titles of the audio tracks. Similar to 'ID,' it is not used as a feature for analysis but can be valuable for identifying tracks.
Measurement Level: Nominal level. 'Name' values are text labels without an inherent order or numerical meaning.

Type (Role: Identifier, Measurement Level: Nominal):
Role: The 'Type' variable denotes the object type of the track. It specifies the type of the track, such as "audio_features."
Measurement Level: Nominal level. 'Type' values are categorical labels without a natural order or numerical interpretation.

```python
# Building the K-means clustering
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# import 4 variables and drop the rest / selecting features for clustering
df_cluster = df[['Energy', 'Loudness', 'Instrumentalness', 'Speechiness']]

# convert df_cluster to matrix
X = df_cluster.values

# scaling - standardizing the data
scaler = StandardScaler()
X = scaler.fit_transform(X)

# random state, we will use 32 (explain why) for a change
# set the random state. different random state seeds might result in different
rs = 32

# defining the number of clusters = 3 and assigned rs, so everytime we run it we get different answers
model = KMeans(n_clusters=3, random_state=rs)
model.fit(X)

# sum of intra-cluster distances
# intra-cluster distances = between clusters, what is the distance
# between different inistiances and centroid. . Inertia is a measure of how tightly packed the clusters are
print("Sum of intra-cluster distance:", model.inertia_)

# 
print("Centroid locations:")
for centroid in model.cluster_centers_:
    print(centroid)

y = model.predict(X)
df_cluster['Cluster_ID'] = y

print("\n----------------------------------------------------------------")
print("Data points assigned into each cluster in variable Energy:")
# data points assigned into each cluster
print("Points for each cluster")
print(df_cluster['Cluster_ID'].value_counts())

print("\n----------------------------------------------------------------")
print("Pairplot for the cluster distribution:")

cluster_plot = sns.pairplot(df_cluster, hue='Cluster_ID')
plt.show()
```

# 7)	How the outcome of the clustering analysis can be used by decision makers? Given an application or an example where this clustering outcome can be used by the music shop.  

```python
from sklearn.metrics import pairwise_distances

# list to save the clusters and cost
clusters = []
inertia_vals = []

# this whole process should take a while
for k in range(2, 15, 2):
    # train clustering with the specified K
    model = KMeans(n_clusters=k, random_state=rs)
    model.fit(X)
    
    # append model to cluster list
    clusters.append(model)
    inertia_vals.append(model.inertia_)

# plot the inertia vs K values
plt.plot(range(2,15,2), inertia_vals, marker='*')
plt.show()

# Compute pairwise distances between cluster centers (using Euclidean distance)
cluster_distances = pairwise_distances(model.cluster_centers_, metric='euclidean')

# Find maximum, minimum, and average distances
max_distance = cluster_distances.max()
min_distance = cluster_distances[cluster_distances > 0].min()  # Exclude 0 distances
average_distance = cluster_distances[cluster_distances > 0].mean()  # Exclude 0 distances

# Print the results
print("Maximum Distance between Cluster Centers:", max_distance)
print("Minimum Distance between Cluster Centers:", min_distance)
print("Average Distance between Cluster Centers:", average_distance)
```
